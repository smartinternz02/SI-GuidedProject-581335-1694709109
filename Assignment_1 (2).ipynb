**TASK - 1**

Create a pandas dataframe (DataFrame name as 'df') with numpy random values (4 features and 4 observation)

```python
import pandas as pd
import numpy as np
```

```python
data = np.random.rand(4, 4)
column = ['F1', 'F2', 'F3', 'F4']
i = ['Inspection1', 'Inspection2', 'Inspection3', 'Inspection4']
df = pd.DataFrame(data, columns=column, index=i)
df
```

**TASK - 2**

Rename the task - 1 'df' dataframe column names to 'Random value 1', 'Random value 2', 'Random value 3' & 'Random value 4'

```python
new_column_names = {
    'F1': 'Random value 1',
    'F2': 'Random value 2',
    'F3': 'Random value 3',
    'F4': 'Random value 4'
}
df.rename(columns=new_column_names, inplace=True)
df
```
